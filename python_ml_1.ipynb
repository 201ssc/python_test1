```python
import numpy as np
x = np.array([[1, 2, 3], [4, 5, 6]])
print("x:\n",x)
```

```python
from scipy import sparse

# 대각선 원소는 1이고 나머지는 0인 2차원 NumPy 배열을 만듭니다.
eye = np.eye(4)
print("NumPy 배열:\n", eye)
```

```python
# NumPy 배열을 CSR 포맷의 SciPy 희박 행렬로 변환합니다.
# 0이 아닌 원소만 저장됩니다.
sparse_matrix = sparse.csr_matrix(eye)
print("\nSciPy의 CSR 행렬:\n", sparse_matrix)
```

```python
x = np.array([[0, 1, 2], [0, 1, 2]])
sx = sparse.csr_matrix(x)
print("\nSciPy의 CSR 행렬:\n", sx)
```

```python
data = np.ones(4)  # > array([1., 1., 1., 1.])
row_indices = np.arange(4)  # > array([0, 1, 2, 3])
col_indices = np.arange(4)  # > array([0, 1, 2, 3])
eye_coo = sparse.coo_matrix((data, (row_indices, col_indices)))  #(array([1., 1., 1., 1.]), (array([0, 1, 2, 3]), array([0, 1, 2, 3])))
print("COO 표현:\n", eye_coo) # ==sparse_matrix = sparse.csr_matrix(eye)
```

```python
import matplotlib.pyplot as plt

# -10에서 10까지 100개의 간격으로 나뉘어진 배열을 생성합니다.
x = np.linspace(-10, 10, 100)
# 사인 함수를 사용하여 y 배열을 생성합니다.
y = np.sin(x)
# plot 함수는 한 배열의 값을 다른 배열에 대응해서 선 그래프를 그립니다.
plt.plot(x, y, marker="x")
plt.show() # 책에는 없음
```
